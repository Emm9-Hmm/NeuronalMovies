```python
import numpy as np
import pandas as pd
```

```python
movies_df = pd.read_csv("imdb_movies_global.csv")
movies_df
```

```python
#secundary genres countries and lang had a space at the beginning
movies_df['genre_2'] = movies_df['genre_2'].str.strip()
movies_df['genre_3'] = movies_df['genre_3'].str.strip()
movies_df['country_2'] = movies_df['country_2'].str.strip()
movies_df['country_3'] = movies_df['country_3'].str.strip()
movies_df['country_4'] = movies_df['country_4'].str.strip()
movies_df['country_5'] = movies_df['country_5'].str.strip()
movies_df['lang_2'] = movies_df['lang_2'].str.strip()
movies_df['lang_3'] = movies_df['lang_3'].str.strip()
movies_df['lang_4'] = movies_df['lang_4'].str.strip()
movies_df['lang_5'] = movies_df['lang_5'].str.strip()
movies_df
```

```python
##transform year to float
movies_df = movies_df.replace("TV Movie 2019", "2019")
movies_df = movies_df.astype({"year_mv": float})
```

```python
c1 = np.concatenate((movies_df['genre_1'].unique(), movies_df['genre_2'].unique()))
c2 = np.concatenate((c1, movies_df['genre_3'].unique()))
list = []
for i in range(len(c2)):
  if (not isinstance(c2[i], str)):
     list.append(i)
list
arr = np.delete(c2, list)
arr = np.unique(arr)
len(arr)
```

```python
c1 = np.concatenate((movies_df['country_1'].unique(), movies_df['country_2'].unique()))
c2 = np.concatenate((c1, movies_df['country_3'].unique()))
c3 = np.concatenate((c2, movies_df['country_4'].unique()))
c4 = np.concatenate((c3, movies_df['country_5'].unique()))
list = []
for i in range(len(c4)):
  if (not isinstance(c4[i], str)):
     list.append(i)
list
arr = np.delete(c4, list)
arr = np.unique(arr)
len(arr)
```

```python
for e in arr:
    #mask = (movies_df['genre_1'] == e) or (movies_df['genre_2'] == e) or (movies_df['genre_3'] == e)
    mask = movies_df['country_1'] == e
    mask2 = movies_df['country_2'] == e
    mask3 = movies_df['country_3'] == e
    mask4 = movies_df['country_4'] == e
    mask5 = movies_df['country_5'] == e
    column_name = 'country_'+e
    movies_df.loc[mask , column_name] = 1
    movies_df.loc[mask2 , column_name] = .8
    movies_df.loc[mask3 , column_name] = .6
    movies_df.loc[mask4 , column_name] = .4
    movies_df.loc[mask5 , column_name] = .2
```

```python
c1 = np.concatenate((movies_df['lang_1'].unique(), movies_df['lang_2'].unique()))
c2 = np.concatenate((c1, movies_df['lang_3'].unique()))
c3 = np.concatenate((c2, movies_df['lang_4'].unique()))
#c4 = np.concatenate((c3, movies_df['lang_5'].unique()))
list = []
for i in range(len(c3)):
  if (not isinstance(c3[i], str)):
     list.append(i)
list
arr = np.delete(c3, list)
arr = np.unique(arr)
len(arr)
arr
```

```python
for e in arr:
    #mask = (movies_df['genre_1'] == e) or (movies_df['genre_2'] == e) or (movies_df['genre_3'] == e)
    mask = movies_df['lang_1'] == e
    mask2 = movies_df['lang_2'] == e
    mask3 = movies_df['lang_3'] == e
    mask4 = movies_df['lang_4'] == e
    #mask5 = movies_df['lang_5'] == e
    column_name = 'lang_'+e
    movies_df.loc[mask , column_name] = 1
    movies_df.loc[mask2 , column_name] = .75
    movies_df.loc[mask3 , column_name] = .50
    movies_df.loc[mask4 , column_name] = .25
    #movies_df.loc[mask5 , column_name] = 1
```
